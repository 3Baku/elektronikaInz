```python
import random as rnd
def volumne_steinmetz_py(N, r):
    rnd.seed(5)
    liczba_trafien = 0
    N=100
    r = 5

    for i in range(0,N):
        x = rnd.random()
        y = rnd.random()
        z = rnd.random()
        if (x**2 + y**2 <= 1) and (x**2 + z**2 <=1):
            liczba_trafien += 1
    V_steinmetza = (2*r)**3 * liczba_trafien/N 
    return(V_steinmetza)

```
